#### Задача 3
Семь человек рассаживаются наудачу на скамейке. Какова вероятность того, что два определённых человека будут сидеть рядом?

#### ОТВЕТ:
Пара на скамейке может сесть 6-ю способами. 
Для каждого из этих способов рассадки вероятность образования нужной пары из группы в 7 человек с учётом того, что не важен порядок следования людей в паре – это Сочетание из 7 по 2 = 7! / (2! * (7 – 2)!) = 7 * 6 / 2 = 21, т.е. нужная пара образуется в 1 случае из 21.

Иными словами, вероятность = 6 * (1 / 21) = 0.28…

#### Задача 5
Бросается игральная кость. Пусть событие A - появление чётного числа, событие B - появление числа больше трёх. Являются ли эти события независимыми?

#### ОТВЕТ:
Чётные числа, которые могут выпасть: 2, 4, 6 (вероятность выпадения ½), 

из них больше 3:  4 и 6 (2 благоприятных исхода из 3, т.е. вероятность = 2/3) => 

вероятность того, что получено число больше трёх при условии, что выпало чётное число = ½ * 1/3 = 1/6

Для независимых событий: Р(B|A) = P(B) или P(AB) = P(A)*P(B), то есть если бы А и В были бы независимыми, то Р(B|A) было бы равно P(B) = 3/6 = ½ или Р(АВ) = P(A)*P(B) = 3/6 * 3/6 = ¼, а это не так.

Т.е. события зависимые.

#### Задача 6
Допустим, имеется некоторая очень редкая болезнь (поражает 0.1% населения). Вы приходите к врачу, вам делают тест на эту болезнь, и тест оказывается положительным. Врач говорит вам, что этот тест верно выявляет 99 % больных этой болезнью и всего лишь в 1% случаев даёт ложный положительный ответ.

Вопрос: какова вероятность, что вы действительно больны ей?
Подсказка: используйте формулу Байеса с раскрытием знаменателя с помощью формулы полной вероятности. 

#### ОТВЕТ: 
Пусть гипотеза 1 (Н1) – болен, гипотеза 2 (Н2) – не болен, тогда полная группа событий Н1 + Н2 => так как Р (Н1)  = 0.001, то Р(Н2) = 0.999

Вероятность правильности теста = 0.99, то есть:
Р(тест положит | H1) = 0.99 и Р(тест положит | H2) = 0.01

Полная вероятность события «тест положительный» Р(тест положит) = Р(тест положит | H1) * Р (Н1)  + Р(тест положит | H2) * Р(Н2) = 0.99 * 0.001 + 0.01 * 0.999 = 0.01097901

Тогда вероятность того, что человек болен при положительном тесте
Р(Н1 | тест положит) = Р(тест положит | H1) * Р(Н1)/Р(тест положит) = 0.99 * 0.001 / 0.01097901 = 0.090172
