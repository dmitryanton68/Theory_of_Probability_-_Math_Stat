$$P(X = 3) = C_n^k p^k (1 - p)^{n - k} = C_5^3 \cdot 0.25^3 \cdot 0.75^2 = \frac{5!}{3! \: (5-3)!} \cdot 0.25^3 \cdot 0.75^2 = 0.087890625 $$

$$P(X = 2) = C_n^k p^k (1 - p)^{n - k} = C_5^2 \cdot 0.25^2 \cdot 0.75^3 = \frac{5!}{2! \: (5-2)!} \cdot 0.25^2 \cdot 0.75^3 = 0.263671875 $$

$$P(X = 1) = C_n^k p^k (1 - p)^{n - k} = C_5^1 \cdot 0.25^1 \cdot 0.75^4 = \frac{5!}{1! \: (5-1)!} \cdot 0.25^1 \cdot 0.75^4 = 0.3955078125 $$

$$P(X = 6) = C_n^k p^k (1 - p)^{n - k} = C_{200}^6 \cdot 0.01^6 \cdot 0.99^{200-6} = \frac{200!}{6! \: (200-6)!} \cdot 0.01^6 \cdot 0.99^{194} $$

$$P(X = 8) = C_n^k p^k (1 - p)^{n - k} = C_{200}^8 \cdot 0.01^8 \cdot 0.99^{200-8} = \frac{200!}{8! \: (200-8)!} \cdot 0.01^8 \cdot 0.99^{192}  $$

$$P(X = 10) = C_n^k p^k (1 - p)^{n - k} = C_{200}^10 \cdot 0.01^{10} \cdot 0.99^{200-10} = \frac{200!}{10! \: (200-5)!} \cdot 0.01^{10} \cdot 0.99^{190} $$

#### Задача 3

В условиях предыдущей задачи сколько выстрелов нужно сделать чтобы быть уверенным с вероятностью 0.9, что хотя бы 10 раз будет совершено попадание?

Подсказка. Как посчитать P(X >= k) для величины X, имеющей распределение Пуассона? Здесь предстоит немножко покодить.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math

%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
n = np.arange(1, 3000)
probability = []
p = 0.01
k = 10

for i in range(len(n)):
    f = (n[i]*p)**k * math.exp(-n[i]*p) / math.factorial(k)
    probability.append(f)
    
plt.plot(n, probability, color='Blue')
```

```python
Proba = 0.9
summ = 0

for i in range(len(probability)):
    summ += probability[i]
    if summ >= Proba:
        limit_num = i
        print(f'Как минимум надо выстрелить {limit_num} раз, чтобы вероятность попадания была более {summ}')
        break
```
